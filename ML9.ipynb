```python
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
```

```python
#get the image refernce from misc modue
face = misc.face()
face.shape
```

```python
face.max()
```

```python
face.dtype
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
#plot the image
plt.gray()
plt.imshow(face)
plt.show()
```

```python
#resahpe the image
face_r = (face / 255.0).reshape(-1,3)
```

```python
#get the new face
face_r.shape
```

```python
from sklearn.cluster import KMeans

#divide and fit into 5 clusters
k_colors = KMeans(n_clusters=5).fit(face_r)
y_pred=k_colors.predict(face_r)
```

```python
#check th new shape of the created cluster
k_colors.cluster_centers_.shape
```

```python
k_colors.labels_.shape
```

```python
newimg=k_colors.cluster_centers_[k_colors.labels_]
```

```python
newimg.shape
```

```python
#reshape the new image as per original image
newimg=np.reshape(newimg, (face.shape))
```

```python
newimg.shape
```

```python
#plot the original and compressed image
fig = plt.figure(figsize=(8,8))
ax=fig.add_subplot(1,2,1,xticks=[],yticks=[],title='Original Image')
ax.imshow(face)
ax=fig.add_subplot(1,2,2,xticks=[],yticks=[],title='Color Compressed Image using K-Means')
ax.imshow(newimg)
plt.show()
```
